```python
import numpy as np
import sympy as sp
from math import atan, acos, asin, sin, cos, sqrt, pi
import matplotlib.pyplot as plt
```

```python
l1=0.042
l2=0.15
alpha_vel = 50
time=np.linspace(0,20,1000000)
theta_1= alpha_vel*time
Y=[]
Theta=[]
```

```python
for theta in theta_1:
    BE=sqrt((l2*l2)-(l1*l1*cos(theta*pi/180)*cos(theta*pi/180)))
    AB=-l1*cos(theta*pi/180)
    #theta_2= atan(BE/AB)
    theta_2=2*pi-asin((l1/l2)*cos(pi*theta/180))
    Ye=l1*sin(pi*theta/180) + l2*cos(pi*theta_2/180)
    Y.append(Ye)
    Theta.append(theta)

plt.plot(Theta,Y)
plt.grid(True)
plt.xlabel("Theta")
plt.ylabel("Position of Piston")
plt.title("Ye v/s Theta")
plt.show()
```

```python
plt.plot(time,Y)
plt.grid(True)
plt.xlabel("Time")
plt.ylabel("Position of Piston")
plt.title("Ye v/s Time")
plt.show()
```

```python
velocity = np.gradient(Y,time) 
```

```python
plt.plot(time,velocity)
plt.grid(True)
plt.xlabel("Time")
plt.ylabel("Velocity of Piston")
plt.title("Velocity of Piston v/s Time")
plt.show()
```

```python
acceleration = np.gradient(velocity,time)
```

```python
plt.plot(Theta,acceleration)
plt.grid(True)
plt.xlabel("Theta")
plt.ylabel("Acceleration of Piston")
plt.title("Accelerationy of Piston v/s Theta")
plt.show()
```
