```python
import sympy as sp
from sympy import cos, sin, atan, acos, asin, sqrt,cot
import numpy as np
import matplotlib.pyplot as plt
pi=3.14
l1 = 0.042
l2 = 0.115
l3 = 0.140
l4 = 0.058
l5 = 0.155
l6 = 0.150
delta = -0.004
Xs = -0.110
Ys = -0.120
ls = 0.030
t = sp.symbols('t')#
omega_vel = 50 #velocity of shaft
theta_1 = omega_vel * t # considering time as variable only
theta_s = 0
```

```python
    Xc = Xs + ls * cos(pi * theta_s / 180)
    Yc = Ys + ls * sin(pi * theta_s / 180)
    Xz = l1 * cos(pi * theta_1 / 180) - Xc
    Yz = l1 * sin(pi * theta_1 / 180) - Yc
    Xz2 = Xz ** 2
    Yz2 = Yz ** 2
    theta_3 = atan(Yz / Xz) - acos((l3 ** 2 - l2 ** 2 + Xz2 + Yz2) / (2 * l3 * sqrt(Xz2 + Yz2)))
    theta_2 = pi - asin((l3 * sin(theta_3) - Yz) / l2)
    theta_4 = theta_2 - acos((l2 ** 2 + l4 ** 2 - l5 ** 2) / (2 * l2 * l4))
    Xd = l1 * cos(pi * theta_1 / 180) + l4 * cos(theta_4)
    Yd = l1 * sin(pi * theta_1 / 180) + l4 * sin(theta_4)
    theta_6 = acos((delta - Xd) / l6)
    Ye = Yd + l6 * cos(theta_6)
```
